```python
from __future__ import absolute_import, division, print_function, unicode_literals

# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
```

```python
fashion_mnist = keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_images_norm = train_images / 255.0

test_images_norm = test_images / 255.0
```

```python
train_images_norm_expand = np.expand_dims(train_images_norm, 3)
```

```python
train_images_norm_expand.shape
```

```python
test_images_norm_expand = np.expand_dims(test_images_norm, 3)
```

```python
test_images_norm_expand.shape
```

```python
# build model 
```

```python
dense_model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
dense_model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
dense_model.fit(train_images_norm, train_labels, epochs=20)
```

```python
test_loss, test_acc = dense_model.evaluate(test_images_norm, test_labels)

print('Test accuracy:', test_acc)
```

```python
# cnn model 
```

```python
cnn_model = keras.Sequential([
    keras.layers.Conv2D(32, kernel_size=3, activation='relu'),
    keras.layers.MaxPooling2D(pool_size=2, strides=2),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
cnn_model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn_model.fit(train_images_norm_expand, train_labels, epochs=20)
```

```python
test_loss, test_acc = cnn_model.evaluate(test_images_norm_expand, test_labels)

print('Test accuracy:', test_acc)
```

```python
# 2 conv 2 pooling cnn
```

```python
cnn_model_2 = keras.Sequential([
    keras.layers.Conv2D(32, kernel_size=3, activation='relu'),
    keras.layers.MaxPooling2D(pool_size=2, strides=2),
    keras.layers.Conv2D(32, kernel_size=3, activation='relu'),
    keras.layers.MaxPooling2D(pool_size=2, strides=2),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
cnn_model_2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn_model_2.fit(train_images_norm_expand, train_labels, epochs=20)
```

```python
test_loss, test_acc = cnn_model_2.evaluate(test_images_norm_expand, test_labels)

print('Test accuracy:', test_acc)
```

```python
# 2 dilated cnn
```

```python
cnn_model_3 = keras.Sequential([
    keras.layers.Conv2D(32, kernel_size=3, activation='relu', dilation_rate=2),
    keras.layers.Conv2D(32, kernel_size=3, activation='relu', dilation_rate=4),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
cnn_model_3.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn_model_3.fit(train_images_norm_expand, train_labels, epochs=20)
```

```python
test_loss, test_acc = cnn_model_3.evaluate(test_images_norm_expand, test_labels)

print('Test accuracy:', test_acc)
```

```python
cnn_model_4 = keras.Sequential([
    keras.layers.Conv2D(8, kernel_size=3, activation='relu', dilation_rate=2),
    keras.layers.Conv2D(8, kernel_size=3, activation='relu', dilation_rate=4),
    keras.layers.Conv2D(8, kernel_size=3, activation='relu', dilation_rate=6),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
cnn_model_4.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn_model_4.fit(train_images_norm_expand, train_labels, epochs=20)
```

```python
test_loss, test_acc = cnn_model_4.evaluate(test_images_norm_expand, test_labels)

print('Test accuracy:', test_acc)
```
